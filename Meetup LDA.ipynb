```python
from gensim import corpora, models
import pandas as pd
from gensim.parsing.preprocessing import STOPWORDS
import gensim
import re
pattern = re.compile(r'"[^"]+"')
```

## First Iteration of LDA

```python
#read data
dat = pd.read_csv('F:\Meetup\Elasticsearch\Articles.csv',encoding='latin').Article.tolist()
stoplist = list(STOPWORDS)
```

```python
#remove stop words
texts = [[word for word in document.lower().split() if word not in stoplist] for document in dat]
#dictionary
dictionary = corpora.Dictionary(texts)
#corpus
corpus = [dictionary.doc2bow(text) for text in texts]
#train model
lda = models.LdaMulticore(corpus, id2word=dictionary, num_topics=25, workers=4,minimum_probability=0)
#print topics
lda.print_topics(num_topics=25, num_words=7)
```

## Third iteration - Bigram LDA

```python
stoplist = list(STOPWORDS)
new = ['said','pakistan','new','world','percent','england','told', 'islamabad', 'india', '--','pakistani','west', 'indies','sri', 'lanka','including','islamabad','islamic','style="text-align:','new', 'zealand','australia','china', 'chinese', 'saudi', 'government', 'international', 'indian', 'expected', 'arabia','<br/>','bangladesh','got','billion','million','u.s.','islamabad']
stoplist.extend(new)
stoplist.sort()
```

```python
texts = [[word for word in document.lower().split() if word not in stoplist] for document in dat]
```

```python
#list of lists
#data_words = 
bigram = gensim.models.Phrases(texts, min_count=5, threshold=100)
```

```python
bigram_mod = gensim.models.phrases.Phraser(bigram)
```

```python
data_words_bigrams = [bigram_mod[doc] for doc in texts if word not in stoplist]
```

```python
dictionary_bigram = corpora.Dictionary(data_words_bigrams)
corpus_bigram = [dictionary_bigram.doc2bow(text) for text in data_words_bigrams]
```

```python
num_top = 180
lda_bigram = models.LdaMulticore(corpus_bigram, id2word=dictionary_bigram, num_topics=num_top, workers=3,minimum_probability=0)
```

```python
#strings_val = []
for i in range(num_top):
    str_val = []
    for i in re.findall(pattern=pattern,string=str(lda_bigram.print_topics(num_topics=num_top, num_words=15)[i])):
        str_val.extend([i.strip('"')])
    print(str_val)
    #print(str(re.findall(pattern=pattern,string=str(lda_bigram.print_topics(num_topics=200, num_words=10)[i]))).strip("'"))
```
